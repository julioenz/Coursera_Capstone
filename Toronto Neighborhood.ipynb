```python
import requests
from lxml import etree
from bs4 import BeautifulSoup

website_url = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M").text
soup = BeautifulSoup(website_url,"lxml")
print(soup.prettify())
```

```python
My_table = soup.find("table",{"class":"wikitable sortable"})
My_table
```

```python
postal = []
borough = []
neigh = []
for row in My_table.findAll("tr"):
    cells = row.findAll('td')
    if len(cells)==3 and "Not assigned" not in cells[1]: 
        postal.append(cells[0].find(text=True))
        borough.append(cells[1].find(text=True))
        neigh.append(cells[2].find(text=True))

    
```

```python
import pandas as pd

df=pd.DataFrame(postal,columns=['Postcode'])
df['Borough']=borough
df['Neighbourhood']=neigh

df
```

```python
df2 = pd.read_csv("Geospatial_Coordinates.csv")
df2
```

```python
df3 = df.merge(df2, on=["Postcode"], how='outer')
df3
```

```python
import folium
latitude = 43.7184038
longitude = -79.5181407

map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(df3['Latitude'],df3['Longitude'], df3['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
